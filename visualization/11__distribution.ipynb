### 4.6 Distrubtion - taqsimot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df['year'] >= 2010] # 2010-yildan oldin ishlab chiqarlgan mashinalrni olamiz
df = df[df['transmission'] != 'Other'] # transmisyasi yozlmagan mashinalrni olib tashlaymiz
df = df[df['fuelType'] != 'Other'] # Yonilg'i turi yozlmagan mashinalrni olib tashlaymiz
df
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(data = df, x = 'price')
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'transmission', multiple = 'stack', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'mileage', multiple = 'stack', hue = 'transmission', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price', col = 'transmission', height = 5, aspect = 2)
plt.show()
```

### 4.6 Distrubtion - taqsimot KDE

```python
sns.displot(data = df, x = 'price', kind = 'kde', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, x = 'price', kde = True, height = 5, aspect = 2)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(data = df, x = 'price', hue = 'fuelType', fill = True)
plt.show()
```
