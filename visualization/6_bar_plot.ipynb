### 4.2. Ustunli grafiklar. sns.barplot()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv")
df.columns = ['viloyat', 'maydoni', 'aholi']
df
```

```python
plt.figure(figsize = (16, 6))



sns.barplot(data = df, x = 'viloyat', y = 'aholi')

plt.title("O'zbekston Respublikasi aholi soni")

plt.xticks(rotation = 90)
plt.grid()

plt.ylabel("Aholi soni")
plt.xlabel("Viloyatlar")

plt.show()
```
