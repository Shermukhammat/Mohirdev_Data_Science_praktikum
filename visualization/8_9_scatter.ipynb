### 8 - daras. `scatter` plot - Tarqoqlik diagrammasi 1-qisim

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df
```

```python
df = df[df['year'] > 2_000]
plt.figure(figsize = (15, 7))

sns.scatterplot(data = df, x = 'year', y = 'price')
sns.regplot(data = df, x = 'year', y = 'price', line_kws={"color": "red"})
plt.show()
```

```python
plt.figure(figsize = (15, 7))

sns.scatterplot(data = df, x = 'mileage', y = 'price')
# sns.regplot(data = df, x = 'mileage', y = 'price', line_kws={"color": "red"})
plt.show()
```

### 9 - daras. `scatter` plot - Tarqoqlik diagrammasi 2-qisim

```python
plt.figure(figsize = (16, 7))

sns.scatterplot(data = df, x = 'mileage', y = 'price', hue = 'transmission', s = 50, style = 'fuelType')
# sns.regplot(data = df, x = 'mileage', y = 'price', line_kws={"color": "red"})
plt.show()
```

```python
plt.figure(figsize = (16, 7))

sns.scatterplot(data = df, x = 'mileage', y = 'price', hue = 'fuelType', s = 50)
# sns.regplot(data = df, x = 'mileage', y = 'price', line_kws={"color": "red"})
plt.show()
```

```python
sns.swarmplot(data = df, x = df['fuelType'], y = df['price'])
plt.show()
```
